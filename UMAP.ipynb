```python
from umap import UMAP
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = load_breast_cancer()
```

```python
data.keys()
```

```python
# lets see description
print(data['DESCR'])
```

```python
# Lets create a dataframe

df = pd.DataFrame(data['data'], columns=data['feature_names'])
df.head()
```

```python
# Here data is in 30d dimension

# Scaling the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
umap_= UMAP(n_neighbors=2, n_components=2)
umap_x = umap_.fit_transform(scaled_data)
```

```python
# Previous dimension
scaled_data.shape
```

```python
# New dimension
umap_x.shape
```

```python
plt.scatter(umap_x[:,0], umap_x[:,1], c=data['target'])
plt.xlabel('PC1')
plt.ylabel('PC2')
```
